##  What is OOP?
OOP (Object-Oriented Programming) organizes code using **classes** and **objects**.

### Four Key Concepts:
1. **Encapsulation** – Bundling data and functions together.
2. **Inheritance** – Reusing code by extending existing classes.
3. **Polymorphism** – Same function name, different behaviors.
4. **Abstraction** – Hiding internal details and showing only the essential parts.

##  1. Creating a Class and Object
Let's start by defining a simple class called `Dog`. A **class** is a blueprint, and an **object** is an instance of that blueprint.

```python
# Example 1: Basic Class and Object

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Creating objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Charlie", 5)

print(dog1.name, dog1.age)
print(dog2.name, dog2.age)
```

##  2. Instance Methods
Instance methods let objects perform actions. Always include `self` as the first argument.

```python
# Example 2: Adding a Method

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def speak(self, sound):
        return f"{self.name} says {sound}"

dog = Dog("Rocky", 4)
print(dog.speak("Woof!"))
```

##  3. Class vs Instance Attributes
- **Instance Attributes:** Belong to each object.
- **Class Attributes:** Shared by all objects of that class.

```python
# Example 3: Class vs Instance Attributes

class Dog:
    species = "Canis familiaris"  # class attribute

    def __init__(self, name, age):
        self.name = name
        self.age = age

dog1 = Dog("Miles", 4)
dog2 = Dog("Buddy", 9)

print(dog1.species)
print(dog2.species)
print(Dog.species)
```

##  4. Inheritance
Inheritance allows one class to use another class’s attributes or methods.

```python
# Example 4: Inheritance

class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "Some sound"

class Dog(Animal):
    def speak(self):
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

dog = Dog("Buddy")
cat = Cat("Kitty")

print(dog.name, "->", dog.speak())
print(cat.name, "->", cat.speak())
```

##  5. Polymorphism
Polymorphism means the same method name behaves differently for different classes.

```python
# Example 5: Polymorphism

def animal_sound(animal):
    print(animal.speak())

animals = [Dog("Max"), Cat("Tom")]
for a in animals:
    animal_sound(a)
```

##  6. Abstraction
Abstraction hides the complex code inside methods and only shows essential features.

```python
# Example 6: Abstraction using Abstract Base Class

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def make_sound(self):
        print("Meow!")

dog = Dog()
cat = Cat()

dog.make_sound()
cat.make_sound()
```

```python
# Encapsulation
class Account:
    def __init__(self, owner, balance):
        self.owner = owner
        self.__balance = balance  # Private variable

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited ₹{amount}. New balance: ₹{self.__balance}")

    def get_balance(self):
        print(f"Current balance: ₹{self.__balance}")

acc = Account("Amit", 5000)
acc.deposit(1500)
acc.get_balance()

# acc.__balance  Can't access directly (Encapsulation hides it)
```
